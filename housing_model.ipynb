```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import pandas as pd

kf = KFold(n_splits=5, shuffle=True, random_state=10)
scaler = StandardScaler()

fold_scores = pd.DataFrame(columns=['Fold', 'Score'])

fold = 1
for train_index, val_index in kf.split(x_train.values):
    x_train_fold, x_val_fold = x_train.values[train_index], x_train.values[val_index]
    y_train_fold, y_val_fold = y_train.values[train_index], y_train.values[val_index]

    x_train_fold_scaled = scaler.fit_transform(x_train_fold)
    x_val_fold_scaled = scaler.transform(x_val_fold)

    regressor = LinearRegression()
    regressor.fit(x_train_fold_scaled, y_train_fold)

    score = regressor.score(x_val_fold_scaled, y_val_fold)

    fold_scores = pd.concat([fold_scores, pd.DataFrame({'Fold': [fold], 'Score': [score]})])

    fold += 1

avg_score = fold_scores['Score'].mean()

fold_scores = pd.concat([fold_scores, pd.DataFrame({'Fold': ['Average Score'], 'Score': [avg_score]})], ignore_index=True)

print(fold_scores)
```

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
import pandas as pd

kf = KFold(n_splits=5, shuffle=True, random_state=10)
scaler = StandardScaler()

fold_scores_knn = pd.DataFrame(columns=['Neighbors', 'Fold', 'Score'])

for n_neighbors in range(1, 16):
    fold = 1
    for train_index, val_index in kf.split(x_train.values):
        x_train_fold, x_val_fold = x_train.values[train_index], x_train.values[val_index]
        y_train_fold, y_val_fold = y_train.values[train_index], y_train.values[val_index]

        x_train_fold_scaled = scaler.fit_transform(x_train_fold)
        x_val_fold_scaled = scaler.transform(x_val_fold)

        knn = KNeighborsRegressor(n_neighbors=n_neighbors)
        knn.fit(x_train_fold_scaled, y_train_fold)

        score_knn = knn.score(x_val_fold_scaled, y_val_fold)

        fold_scores_knn = pd.concat([fold_scores_knn, pd.DataFrame({'Neighbors': [n_neighbors], 'Fold': [fold], 'Score': [score_knn]})])

        fold += 1

avg_scores = fold_scores_knn.groupby('Neighbors')['Score'].mean().reset_index()
avg_scores.columns = ['Neighbors', 'Average Score']

print("\nAverage Scores for each Neighbors:")
print(avg_scores)
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

kf = KFold(n_splits=5, shuffle=True, random_state=10)

fold_scores_rf = pd.DataFrame(columns=['n_estimators', 'Fold', 'Score'])

for n_estimators in [50, 100, 200, 300]:
    fold = 1
    for train_index, val_index in kf.split(x_train.values):
        x_train_fold, x_val_fold = x_train.values[train_index], x_train.values[val_index]
        y_train_fold, y_val_fold = y_train.values[train_index], y_train.values[val_index]

        regressor_rf = RandomForestRegressor(n_estimators=n_estimators, random_state=10)

        regressor_rf.fit(x_train_fold, y_train_fold)

        score_rf = regressor_rf.score(x_val_fold, y_val_fold)

        fold_scores_rf = pd.concat([fold_scores_rf, pd.DataFrame({'n_estimators': [n_estimators], 'Fold': [fold], 'Score': [score_rf]})])

        fold += 1

    average_score_rf = fold_scores_rf[fold_scores_rf['n_estimators'] == n_estimators]['Score'].mean()
    fold_scores_rf = pd.concat([fold_scores_rf, pd.DataFrame({'n_estimators': [n_estimators], 'Fold': ['Average Score'], 'Score': [average_score_rf]})])

print("Random Forest Fold Scores:")
print(fold_scores_rf)
```
